```python
# exercice 2


from PIL import Image



im = Image.new("RGB",(800,600),"lightgrey")
(l, h)= im.size
im.putpixel((200,100),(0,0,0))
im.putpixel((500,200),(0,0,0))
im.putpixel((600,500),(0,0,0))
im.putpixel((700,400),(0,0,0))
im.putpixel((500,100),(0,0,0))
for x in range(l):
    im.putpixel((x,250),(0,0,0))
for y in range(h):
    im.putpixel((250,y),(0,0,0))
for y in range(0,h,60):
    for x in range(0,l):
        im.putpixel((x,y),(0,0,255))
for x in range(0,l,60):
    for y in range(0,h):
        im.putpixel((x,y),(0,255,0))
for y in range(100,300):
    for x in range(500,700):
        im.putpixel((x,y),(255,255,0))
for x in range(0,30):
     for y in range(0,30):   
        for i in range(30):
            for j in range(30):
                im.putpixel((x+i,y+j),(255,0,255))

        
im.show()

```

```python
from PIL import Image, ImageDraw

# Dimensions de l'image
largeur_image = 800
hauteur_image = 600

# Couleurs
couleur_fond = (169, 169, 169)  # Gris
couleur_carre = (138, 43, 226)  # Violet

# Taille et espacement des carrés
taille_carre = 30
espacement_carre = 30

# Création de l'image avec fond gris
image = Image.new("RGB", (largeur_image, hauteur_image), couleur_fond)

# Obtenir le contexte de dessin
dessin = ImageDraw.Draw(image)

# Dessiner les carrés violets
for x in range(0, largeur_image, taille_carre + espacement_carre):
    for y in range(0, hauteur_image, taille_carre + espacement_carre):
        dessin.rectangle(
            [x, y, x + taille_carre, y + taille_carre],
            fill=couleur_carre,
            outline=None  # Si vous voulez un contour, vous pouvez spécifier une couleur ici
        )

# Enregistrement de l'image
image.save("image_resultat.png")

# Affichage de l'image
image.show()
```
