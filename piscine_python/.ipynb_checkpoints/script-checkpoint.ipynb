```python
import os
import glob
import nbformat
from pprint import pprint

# Dictionnaire pour stocker les valeurs des variables trouvées
valeurs_variables = {}
dico_eleve= {}




def listdirectory(path,n): 
    extension = '*.ipynb'
    fichiers_ipynb = [fichier for fichier in os.listdir(path) if fichier.endswith('.ipynb')]
    for noms in fichiers_ipynb:
        nom = noms.split(".")[0]
        if noms != "script.ipynb":
            with open(path + noms, 'r',encoding='utf-8') as fichier:
                contenu = nbformat.read(fichier,as_version=4)
                for cellule in contenu['cells'][:-2]:
                    if cellule['cell_type'] == 'code':
                        # Extraire le contenu du code dans la cellule
                        code = cellule['source']
                        for liste in n:
                            if liste in code:
                                exec(code)  # Exécute le code de la cellule
                                valeurs_variables[liste] = locals()[liste]
                                break
                # Afficher les valeurs des variables trouvées
                for note, valeur in valeurs_variables.items():
                    if nom not in dico_eleve:
                        dico_eleve[nom] = {}
                    dico_eleve[nom][note]= valeur
    pprint(dico_eleve,indent=4)


    
    
            
# Nom de la variable à rechercher
note = ["note00", "note01", "note1", "note2", "note3", "note4", "note5"]          
chemin = r"C:\Users\ludovic.souquet\Documents\GitHub\ludovic.souquet\piscine_python\Evaluation python classroom ciel 28092923\Evaluation python classroom ciel 28092923\\"

listdirectory(chemin, note)
```
